```python
import pandas as pd
import glob
import random
import os
```

```python
# KT2 데이터 경로 설정
data_path = "C:/Users/User/python/project/EdNet-KT2/KT2/u*.csv"

# 1️.전체 파일 불러오기
all_files = glob.glob(data_path) # 경로에 해당하는 모든 파일명 리스트 반환

# 2️.5000명 샘플링
sample_files = random.sample(all_files, 5000)

# 3.user_id 변수 생성(학생 데이터 구분) + 학생 데이터 통합
first = True

for file in sample_files:
    user_id = os.path.basename(file).split('.')[0]

    df = pd.read_csv(file)
    df['user_id'] = user_id

    if first:
        df.to_csv('kt2_5000_users.csv', index=False, mode='w')
        first = False
    else:
        df.to_csv('kt2_5000_users.csv', index=False, mode='a', header=False)
```

```python
# CSV 파일 DB에 적재
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:090909@localhost/project")

df = pd.read_csv("kt2_5000_users.csv")

df.to_sql(name='kt2_log', con=engine, if_exists='replace', index=False)
```
